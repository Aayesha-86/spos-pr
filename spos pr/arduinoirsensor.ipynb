```python
/*
  Roll No: TB - 54
  Title: IR Sensor Based Output Control
  Description:
  This program reads input from an IR sensor connected to analog pin A0.
  When the IR sensor detects an object (HIGH signal), it turns ON an LED
  connected to digital pin 13. When no object is detected (LOW signal),
  the LED turns OFF.
*/

void setup()
{
  // Initialize serial monitor (optional for debugging)
  Serial.begin(9600);

  // Configure IR sensor input pin
  pinMode(A0, INPUT); // IR sensor connected to A0

  // Configure output pins
  pinMode(0, OUTPUT);
  pinMode(1, OUTPUT);
  pinMode(2, OUTPUT);
  pinMode(4, OUTPUT);
  pinMode(13, OUTPUT); // LED or output indicator

  // Initialize all outputs to LOW
  digitalWrite(0, LOW);
  digitalWrite(1, LOW);
  digitalWrite(2, LOW);
  digitalWrite(4, LOW);
  digitalWrite(13, LOW);
}

void loop()
{
  // Read IR sensor state
  int sensorState = digitalRead(A0);

  // Display sensor state on Serial Monitor
  Serial.print("IR Sensor State: ");
  Serial.println(sensorState);

  // If IR sensor detects an object (HIGH)
  if (sensorState == HIGH)
  {
    digitalWrite(13, HIGH); // Turn ON LED
    // You can also activate other outputs if needed
    // digitalWrite(4, HIGH);
  }
  else
  {
    digitalWrite(13, LOW); // Turn OFF LED
    // digitalWrite(4, LOW);
  }

  delay(100); // Small delay to stabilize reading
}
```
