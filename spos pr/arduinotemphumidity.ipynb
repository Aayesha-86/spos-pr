```python
/*
  Roll No: TB - 54
  Title: DHT11 Temperature and Humidity Sensor
  Description:
  This program reads temperature and humidity values from a DHT11 sensor
  connected to digital pin 4 and displays the readings on the Serial Monitor
  every 2 seconds.
*/

#include <SD.h>        // Included but not used here; can be removed if SD card is not used
#include "DHT.h"       // Include DHT sensor library

#define DHTPIN 4       // DHT11 sensor connected to digital pin 4
#define DHTTYPE DHT11  // Define the type of DHT sensor

// Create DHT object
DHT dht(DHTPIN, DHTTYPE);

void setup() {
  // Start serial communication
  Serial.begin(9600);
  Serial.println("=== Humidity and Temperature Monitoring ===");

  // Initialize DHT sensor
  dht.begin();
}

void loop() {
  // Wait a few seconds between readings
  delay(2000);

  // Read humidity and temperature (in Â°C)
  float humidity = dht.readHumidity();
  float temperature = dht.readTemperature();

  // Check if readings failed
  if (isnan(humidity) || isnan(temperature)) {
    Serial.println("Failed to read from DHT sensor!");
    return;
  }

  // Display results on Serial Monitor
  Serial.print("\nHumidity: ");
  Serial.print(humidity);
  Serial.println(" %");

  Serial.print("Temperature: ");
  Serial.print(temperature);
  Serial.println(" *C");
}
```
